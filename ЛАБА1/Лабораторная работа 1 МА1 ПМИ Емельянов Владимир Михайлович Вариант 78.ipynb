Предположим, у Вас есть начальная сумма денег `initial`, и Вы хотите инвестировать ее под фиксированный годовой процентный доход `extra`. Вам нужно найти, сколько времени `time` потребуется, чтобы Ваша инвестиция достигла или превысила определенную целевую сумму `final`.

 Уравнение для этой задачи выглядит следующим образом:

$$ initial × \left(1 + \frac{extra}{100}\right)^{time} = final $$

Напишите функцию `check_time(initial: float, extra: float, final: float) -> int` чтобы определить сколько времени понадобится чтобы достичь желаемой цели.

```python
def check_time(initial, extra, final) -> int:
    time = 0
    current_amount = initial
    
    while current_amount < final:
        current_amount *= (1 + extra / 100)
        time += 1
    
    return time

print(check_time(10000, 12, 50000))
```

**<font color=#FF5EDD>Задача 9**</font> (3 балла)

Вам дана кривая `f_9`. Найдите точку на этой кривой, которая будет находиться на минимальном расстоянии от точки (`x_0`, `y_0`).

```python
x = Symbol('x')



def closest_point(f_9, x_0, y_0):

  x = sp.symbols('x')
  
  y = f_9
  
  # уравнение растояния
  distance = sp.sqrt((x - x_0)**2 + (y - y_0)**2)
  

  distance_derivative = sp.diff(distance, x)
  
  # находим точки экстремумумов у функции distance
  critical_points = [x for x in sp.solve(Eq(distance_derivative, 0), x) if x.is_real]

  # выбираем минимальный экстремум distance, то есть минимальное растояние от (x, y) до (x_0, y_0)
  minn = min(critical_points, key=lambda point: distance.subs(x, point))

  return (minn, f_9.subs(x, minn))

from sympy import Equivalent
```

```python
#TEST CASE

try:
  assert Equivalent(closest_point(x**2 + 2*x + 3, 2, 2), (0, 3))
  print("GOOD")
except:
  print("BAD")
```

Вам дана функция `f_11`. Найдите уравнение касательной к `f_11` в точке `x_11`.

```python
x = Symbol('x')

def GetTangent(f_11, x_11):
  df_11 = sp.diff(f_11, x)
  k = df_11.subs(x, x_11)

  y_11 = f_11.subs(x, x_11)

  b = y_11 - k*x_11

  return k*x + b
```
